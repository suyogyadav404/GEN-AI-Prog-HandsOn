```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

##### -------------------------------------------------------------------------------------------

Import matplotlib.pyplot as plt and set %matplotlib inline

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

##### -------------------------------------------------------------------------------------------

#### Exercise 1

Follow along with these steps:Create a figure object called fig using plt.figure()Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig = plt.figure()
 
ax = fig.add_axes([0,0,1,1])
 
ax.plot(x,y)
 
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
 
plt.show()
```

##### -------------------------------------------------------------------------------------------

Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits onthe inserted plot:

```python
fig = plt.figure(figsize=(8,6))
 
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])
 
# Main plot
ax1.plot(x,z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
 
# Inset plot
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
 
plt.show()
```

Use plt.subplots(nrows=1, ncols=2) to create the plot below.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.show()
```

##### -------------------------------------------------------------------------------------------

Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
 
axes[0].plot(x, y, color='blue', lw=2, ls='--')
axes[0].set_title('x vs y')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
 
axes[1].plot(x, z, color='red', lw=3)
axes[1].set_title('x vs z')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
 
plt.tight_layout()
plt.show()
plt.show
 
```
