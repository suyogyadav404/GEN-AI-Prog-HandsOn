```python
# 4. Print all keys and values
print("Printing all keys and values:")
for key, value in student_data.items():
    if isinstance(value, dict):
        print(f"{key}:")
        for subject, score in value.items():
            print(f"  . {subject}: {score}")
    else:
        print(f"{key}: {value}")
 
```

8. Shallow vs Deep Copy 

• Create a nested list 

• Make a shallow copy and modify an inner element 

• Make a deep copy and modify an inner element 

• Print all lists and explain the difference 

```python
import copy
 
# 1. Create a nested list
original_list = [[1, 2], [3, 4]]

original_list
```

```python
# 2. Make a shallow copy and modify an inner element
shallow_copied_list = copy.copy(original_list)
shallow_copied_list[0][0] = 100

print(f"Original List: {original_list}")
print(f"Shallow Copy:  {shallow_copied_list}")
```

```python
original_list = [[1, 2], [3, 4]]
 
# 3. Make a deep copy and modify an inner element
deep_copied_list = copy.deepcopy(original_list)
deep_copied_list[0][0] = 100
 
print(f"Original List: {original_list}")
print(f"Deep Copy:     {deep_copied_list}")
 
```

Difference between Shallow Copy and Deep Copy

Shallow copy creates a new object but shares references to nested objects, whereas deep copy creates a completely independent copy including all nested objects. Changes in nested objects affect both copies in shallow copy, but not in deep copy.

9. Using Logical and Bitwise Operators 

• Use and, or, and not with some conditions 

• Use &, |, ^, ~, <<, and >> with integers 

• Explain the output of each operation 

```python
# 1. Use and, or, and not with some conditions

a = 10
b = 5

print("AND Operation:", a > 5 and b > 2)      # True: Both conditions are True
print("OR Operation:", a > 15 or b > 2)       # True: One condition is True
print("NOT Operation:", not(a > 5))           # False: Reverses True to False
```

```python
# 2. Use &, |, ^, ~, <<, and >> with integers

x = 10    # Binary: 1010
y = 6     # Binary: 0110

print("Bitwise AND (&):", x & y)              # 2  -> Common 1 bits (0010)
print("Bitwise OR (|):", x | y)               # 14 -> Combines all 1 bits (1110)
print("Bitwise XOR (^):", x ^ y)              # 12 -> 1 where bits are different (1100)
print("Bitwise NOT (~):", ~x)                 # -11 -> Inverts all bits
print("Left Shift (<<):", x << 2)             # 40 -> Shifts left by 2 bits (10 × 4)
print("Right Shift (>>):", x >> 1)            # 5  -> Shifts right by 1 bit (10 ÷ 2)
```

10. Looping Through Data Structures 

• Loop through a list and print each item 

• Loop through a dictionary and print keys, values, and items 

• Loop through a list of tuples and unpack each tuple 

```python
# 1. Loop through a list and print each item

fruits = ["Apple", "Banana", "Mango", "Orange"]

print("List Items:")
for fruit in fruits:
    print(fruit)    
```

```python
# 2. Loop through a dictionary and print keys, values, and items

student = {
    "Name": "Suyog",
    "Age": 26,
    "City": "Mumbai"
}

print("Dictionary Keys:")
for key in student.keys():
    print(key)       

print("\nDictionary Values:")
for value in student.values():
    print(value)     

print("\nDictionary Items:")
for key, value in student.items():
    print(key, ":", value)  
```

```python
# 3. Loop through a list of tuples and unpack each tuple

students = [
    ("Suyog", 26),
    ("Rahul", 23),
    ("Priya", 22)
]

print("List of Tuples:")
for name, age in students:
    print("Name:", name, ", Age:", age) 
```
