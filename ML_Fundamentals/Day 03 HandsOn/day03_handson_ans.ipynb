#### Assignment:  Sales forecast based on Advertising Spend using Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# Load the CSV dataset
df = pd.read_csv('advertising_spend.csv')
 
# Handle any column truncation or spelling issues dynamically
df.columns = ['Advertising_Spend', 'Sales_Revenue']
 
# Display the first 5 rows to verify
df.head()
```

```python
# Separate features (X) and target variable (y)
X = df[['Advertising_Spend']]
y = df['Sales_Revenue']
 
# Initialize and fit the model
model = LinearRegression()
model.fit(X, y)
 
# Generate predictions for the dataset
y_pred = model.predict(X)
 
print("Model training complete.")
print(f"Intercept (w0): {model.intercept_:.4f}")
print(f"Slope (w1): {model.coef_[0]:.4f}")
```

```python
df_predictions = df.copy()
df_predictions['Predicted_Sales_Revenue'] = y_pred
 
print("--- Data Predictions ---")
print(df_predictions.head())
print("\n" + "="*50 + "\n")
 
sample_spend = 7500.00  
new_prediction = model.predict([[sample_spend]])
 
print(f"--- Custom Scenario Prediction ---")
print(f"If you spend: ${sample_spend:,.2f} on advertising,")
print(f"The predicted product sales revenue is: ${new_prediction[0]:,.2f}")
```

b.  Generate a heat map of Advertising Spend Vs Sales Revenue

```python
# Calculate correlation matrix
correlation_matrix = df.corr()
 
# Generate the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".4f", cbar=True)
plt.title('Heatmap: Advertising Spend Vs Sales Revenue')
plt.tight_layout()
plt.show()
```

c.  Visualize the data as a scatter plot with a regression line

```python
plt.figure(figsize=(8, 5))
 
# Plot the real data points
plt.scatter(X, y, color='royalblue', alpha=0.8, edgecolor='k', label='Actual Data')
 
# Plot the calculated regression line
plt.plot(X, y_pred, color='crimson', linewidth=2, label='Regression Line')
 
# Labels and styling
plt.title('Sales Revenue Forecast based on Advertising Spend')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Sales Revenue ($)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

d.  Compute Mean Squared Error, Absolute Mean Error & R2 Score 

```python
# Compute metrics
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)
 
# Display results cleanly
print("=========================================")
print("          EVALUATION METRICS             ")
print("=========================================")
print(f"Mean Squared Error (MSE)      : {mse:.4f}")
print(f"Mean Absolute Error (MAE)     : {mae:.4f}")
print(f"R2 Score (Variance Explained) : {r2:.4f}")
print("=========================================")
 
```
